A counter is created in the code below to put 30 images in each category and the rest for testing data.

```python
#setting empty arrays for training and testing sets
x_train = []
y_train = []
x_test = []
y_test = []

for category in os.listdir(path):
    count = 0
    elements_in_category = os.listdir(os.path.join(path, category))
    #Shuffle images in each category
    random.shuffle(elements_in_category)
    for element in elements_in_category:
        fullpath = os.path.join(path, category, element)        
        label = fullpath.split(os.path.sep)[-2]
        image = cv2.imread(fullpath)
        image = cv2.resize(image, (dimh, dinw))
        if (count < 30):      
            x_train.append(image)
            y_train.append(label)
        else:
            x_test.append(image)
            y_test.append(label)
        count = count + 1
```

```python
from sklearn.preprocessing import LabelBinarizer

x_train = np.array(x_train, dtype="float") / 255.0
x_test = np.array(x_test, dtype="float") / 255.0
lb = LabelBinarizer()
y_train = lb.fit_transform(y_train)
y_test = lb.fit_transform(y_test)
```

```python
import tensorflow.python.keras
from tensorflow.python.keras.layers import Conv2D, MaxPooling2D, ZeroPadding2D, Convolution2D, Dense, Dropout, Activation, Flatten, BatchNormalization
import matplotlib.pyplot as plt
from tensorflow.python.keras.datasets import cifar10
from tensorflow.python.keras.preprocessing.image import ImageDataGenerator
from tensorflow.python.keras.layers import GaussianNoise
from keras import optimizers
from tensorflow.python.keras.models import Sequential
from tensorflow.python.keras.layers.core import Lambda
from tensorflow.python.keras import regularizers
from tensorflow.keras.callbacks import LearningRateScheduler
#import tensorflow as tf
import numpy as np

vgg16model = Sequential()
image_res = [200,200,3]
wg_decay = 0.0005

vgg16model.add(Conv2D(64, (3, 3), padding='same', input_shape=image_res, kernel_regularizer=regularizers.l2(wg_decay)))
vgg16model.add(Activation('relu'))
vgg16model.add(BatchNormalization())
vgg16model.add(Dropout(0.3))

vgg16model.add(Conv2D(64, (3, 3), padding='same',kernel_regularizer=regularizers.l2(wg_decay)))
vgg16model.add(Activation('relu'))
vgg16model.add(BatchNormalization())

vgg16model.add(MaxPooling2D(pool_size=(2, 2)))

```

```python

vgg16model.add(Flatten())
vgg16model.add(Dense(256, activation='relu', name = 'fc1'))
vgg16model.add(Dense(256, activation='relu', name = 'fc2'))
vgg16model.add(BatchNormalization())

vgg16model.add(Dropout(0.5))
vgg16model.add(Dense(102))
vgg16model.add(Activation('softmax'))
```

```python
learning_rate = 1e-4
learning_decay = 1e-4 / 50
learning_drop = 20

Adam = tensorflow.keras.optimizers.Adam(lr=learning_rate, beta_1=0.9, beta_2=0.999, amsgrad=False)

vgg16model.compile(loss='categorical_crossentropy', optimizer=Adam, metrics=['accuracy', 'mse'])
```

```python
from matplotlib import pyplot

vgg16data = ImageDataGenerator(
    #rotation_range=15,
    #horizontal_flip=False,
    #width_shift_range=0.1,
    #height_shift_range=0.1,
    featurewise_center=True,
    #featurewise_std_normalization=True,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
    #zoom_range=0.3
    )
vgg16data.fit(x_train)


```

```python
def lr_scheduler(epoch):
            return learning_rate * (0.5 ** (epoch // learning_drop))
    
reduce_lr = tensorflow.keras.callbacks.LearningRateScheduler(lr_scheduler)
```

```python

history=vgg16model.fit_generator(vgg16data.flow(x_train, y_train, batch_size=128),
                       #steps_per_epoch = len(x_train) / 64,
                       steps_per_epoch = 400, 
                       epochs=40, 
                       validation_data= (x_test, y_test),verbose=1) 
                       #callbacks=[reduce_lr], 
```

```python
def plotAcc(history):  
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title('model Accuracy')
    plt.ylabel('Accuracy')
    plt.xlabel('epoch')
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()
```

```python
plotAcc(history)
```

```python
def plotLosses(history):  
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title('model loss')
    plt.ylabel('loss')
    plt.xlabel('epoch')
    plt.legend(['train', 'validation'], loc='upper left')
    plt.show()
```

```python
plotLosses(history)
```
